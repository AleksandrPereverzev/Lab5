```python
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
from matplotlib.ticker import FuncFormatter
import re

sns.set_theme(style="whitegrid", context="notebook")

def abbreviate(name):
    """Abbreviate indicator names using predefined mapping"""
    mapping = {
        "adjusted net national income per capita (annual % growth)": "ANNIPCA",
        "adjusted net national income per capita (constant 2015 us$)": "ANNIPCC",
        "adjusted net national income (annual % growth)": "ANNIAG",
        "adjusted net national income (constant 2015 us$)": "ANNIC",
        "adjusted net national income (current us$)": "ANNICUR",
    }
    return mapping.get(name.strip().lower(), "Indicator") if isinstance(name, str) else "Indicator"

# Prepare data
years = sorted(int(c) for c in df_russia.columns if str(c).isdigit())
cols = [str(y) for y in years]
df_plot = df_russia.copy()
df_plot[cols] = df_plot[cols].apply(pd.to_numeric, errors="coerce")

# Get name column
name_col = next((c for c in ["Indicator Name", "Series Name", "Indicator"] if c in df_plot.columns), None)

# Select indicators
target_names = [
    "Adjusted net national income (constant 2015 US$)",
    "Adjusted net national income (current US$)",
]
subset = pd.DataFrame([df_plot[df_plot[name_col] == t].dropna(how="all", subset=cols).iloc[0] for t in target_names])
print(f"Number of indicators: {len(subset)}")

# Create plot
fig, ax = plt.subplots(figsize=(10, 4))
for i, (_, row) in enumerate(subset.iterrows()):
    y = row[cols].to_numpy(dtype=float)
    label = abbreviate(row[name_col])
    ax.plot(years, y, marker="o", linewidth=2, label=label, alpha=0.8)
    if not np.isnan(y[-1]):
        ax.text(years[-1] + 0.3, y[-1], label, fontsize=9, va="center", fontweight='bold')

ax.set_title(f"Indicator trends ({years[0]}–{years[-1]})", loc="left")
ax.set_xlabel("Year")
ax.set_ylabel("Value (US$)")
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, _: f"{v:,.0f}"))
ax.legend(loc='best')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import re

sns.set_theme(style="whitegrid", context="notebook")

def abbreviate_indicator(name):
    """Abbreviate long indicator names for better display"""
    if not isinstance(name, str):
        return "Indicator"
    name = re.sub(r'\(constant 2015 US\$\)', '(const.)', name, flags=re.IGNORECASE)
    name = re.sub(r'\(current US\$\)', '(curr.)', name, flags=re.IGNORECASE)
    return re.sub(r'Adjusted net national income', 'ANNI', name, flags=re.IGNORECASE)

last_year = str(sorted([int(c) for c in df_russia.columns if str(c).isdigit()])[-1])

# Prepare data
df_temp = df_russia.copy()
df_temp[last_year] = pd.to_numeric(df_temp[last_year], errors='coerce')
df_filtered = df_temp[~df_temp['Series Name'].str.contains('growth|%|per capita', case=False, na=False)]

# Get top 2 indicators
top = df_filtered[['Series Name', last_year]].nlargest(2, last_year).iloc[::-1]
top['Series Name'] = top['Series Name'].apply(abbreviate_indicator)

# Create plot
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=top, y='Series Name', x=last_year, palette='viridis', ax=ax)

ax.set_title(f"Russia — Comparison: Current vs Constant National Income ({last_year})", loc='left', fontsize=14, fontweight='bold')
ax.set_xlabel("Value (US$)", fontsize=11)
ax.set_ylabel("")
ax.xaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
ax.bar_label(ax.containers[0], fmt='{:,.0f}', padding=3, fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns

sns.set_theme(style="whitegrid", context="notebook")

# Prepare data
last_year = str(sorted([int(c) for c in df_russia.columns if str(c).isdigit()])[-1])
df_temp = df_russia.copy()
df_temp[last_year] = pd.to_numeric(df_temp[last_year], errors='coerce')

summary = pd.DataFrame({'Indicator': df_temp['Series Name'], f'Value ({last_year})': df_temp[last_year]})
summary['Type'] = summary['Indicator'].apply(
    lambda x: 'Percentage/Growth' if any(w in x.lower() for w in ['growth', '%']) else 'Absolute Value (US$)'
)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Dollar indicators
dollar_data = summary[summary['Type'] == 'Absolute Value (US$)'].sort_values(f'Value ({last_year})')
ax1.barh(range(len(dollar_data)), dollar_data[f'Value ({last_year})'], color=sns.color_palette("Blues_r", len(dollar_data)))
ax1.set_yticks(range(len(dollar_data)))
ax1.set_yticklabels(['Total (curr.)' if 'current' in ind.lower() and 'per capita' not in ind.lower() else
                      'Total (const.)' if 'constant' in ind.lower() and 'per capita' not in ind.lower() else
                      'Per Capita (const.)' for ind in dollar_data['Indicator']], fontsize=10)
ax1.set_xlabel("Value (US$)", fontsize=11)
ax1.set_title("Absolute Values", fontsize=12, fontweight='bold')
ax1.set_xscale('log')
ax1.set_xticks([])
ax1.grid(False)
for i, val in enumerate(dollar_data[f'Value ({last_year})']):
    ax1.text(val / 2, i, f'${val/1e12:.2f}T' if val >= 1e12 else f'${val/1e3:.1f}K',
             va='center', ha='center', fontsize=11, color='black', fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8, edgecolor='none'))

# Plot 2: Percentage indicators
pct_data = summary[summary['Type'] == 'Percentage/Growth'].sort_values(f'Value ({last_year})')
ax2.barh(range(len(pct_data)), pct_data[f'Value ({last_year})'], color=sns.color_palette("Reds_r", len(pct_data)))
ax2.set_yticks(range(len(pct_data)))
ax2.set_yticklabels(['Total Growth' if 'per capita' not in ind.lower() else 'Per Capita Growth'
                      for ind in pct_data['Indicator']], fontsize=10)
ax2.set_xlabel("Growth Rate", fontsize=11)
ax2.set_title("Growth Rates", fontsize=12, fontweight='bold')
ax2.set_xticks([])
ax2.grid(False)
for i, val in enumerate(pct_data[f'Value ({last_year})']):
    ax2.text(val / 2, i, f'{val:.1f}%', va='center', ha='center', fontsize=10, color='white', fontweight='bold')

fig.suptitle(f"Russia — Economic Indicators Summary ({last_year})", fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="notebook")

years_str = [str(y) for y in sorted([int(c) for c in df_russia.columns if str(c).isdigit()])]

df_temp = df_russia.copy()
df_temp[years_str] = df_temp[years_str].apply(pd.to_numeric, errors='coerce')

# Extract indicator data
current_income = df_temp[df_temp['Series Name'].str.contains('current', case=False, na=False) &
                         ~df_temp['Series Name'].str.contains('per capita', case=False, na=False)][years_str].values.flatten()
constant_income = df_temp[df_temp['Series Name'].str.contains('constant', case=False, na=False) &
                          ~df_temp['Series Name'].str.contains('per capita', case=False, na=False)][years_str].values.flatten()
total_growth = df_temp[df_temp['Series Name'].str.contains('growth', case=False, na=False) &
                       ~df_temp['Series Name'].str.contains('per capita', case=False, na=False)][years_str].values.flatten()
per_capita_growth = df_temp[df_temp['Series Name'].str.contains('per capita.*growth', case=False, na=False, regex=True)][years_str].values.flatten()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Current vs Constant Income
mask1 = ~(np.isnan(current_income) | np.isnan(constant_income))
x1, y1 = constant_income[mask1], current_income[mask1]
sns.scatterplot(x=x1, y=y1, ax=ax1, s=100, alpha=0.7, color='steelblue', label='Yearly Data')
sns.regplot(x=x1, y=y1, ax=ax1, scatter=False, color='red', line_kws={'linewidth': 2}, label='Trend Line')
r1 = np.corrcoef(x1, y1)[0, 1]
ax1.set_title(f"Current vs Constant Income\n(r = {r1:.3f})", fontsize=13, fontweight='bold')
ax1.set_xlabel("Constant 2015 US$ (Trillions)", fontsize=11)
ax1.set_ylabel("Current US$ (Trillions)", fontsize=11)
ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e12:.1f}T'))
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e12:.1f}T'))
ax1.grid(True, alpha=0.3)
ax1.legend(loc='upper left', fontsize=10, frameon=True, shadow=True)

# Plot 2: Total Growth vs Per Capita Growth
mask2 = ~(np.isnan(total_growth) | np.isnan(per_capita_growth))
x2, y2 = total_growth[mask2], per_capita_growth[mask2]
sns.scatterplot(x=x2, y=y2, ax=ax2, s=100, alpha=0.7, color='coral', label='Yearly Data')
sns.regplot(x=x2, y=y2, ax=ax2, scatter=False, color='darkred', line_kws={'linewidth': 2}, label='Trend Line')
r2 = np.corrcoef(x2, y2)[0, 1]
ax2.set_title(f"Total vs Per Capita Growth\n(r = {r2:.3f})", fontsize=13, fontweight='bold')
ax2.set_xlabel("Total Growth (%)", fontsize=11)
ax2.set_ylabel("Per Capita Growth (%)", fontsize=11)
ax2.axhline(y=0, color='gray', linestyle='--', linewidth=1, alpha=0.5, label='Zero Line')
ax2.axvline(x=0, color='gray', linestyle='--', linewidth=1, alpha=0.5)
ax2.grid(True, alpha=0.3)
ax2.legend(loc='upper left', fontsize=10, frameon=True, shadow=True)

fig.suptitle(f"Russia — Indicator Correlations Across All Years", fontsize=15, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()

print(f"\nCorrelation Analysis:")
print(f"1. Current vs Constant Income: r = {r1:.3f} ({len(x1)} data points)")
print(f"2. Total vs Per Capita Growth: r = {r2:.3f} ({len(x2)} data points)")
```
